```python
import pandas as pd

# Load your CSV file
data = pd.read_csv("weather data.csv")

# Show first 5 rows
print(data.head())
```

```python
# Shape of data (rows, columns)
print("Shape of dataset:", data.shape)

# Column names and data types
print("\nInfo about dataset:")
print(data.info())

# Quick stats (mean, min, max, etc.)
print("\nStatistics:")
print(data.describe())

# Check for missing values
print("\nMissing values:\n", data.isnull().sum())
```

```python
import numpy as np

# Average, max, min temperature
avg_temp = np.mean(data['Temperature'])
max_temp = np.max(data['Temperature'])
min_temp = np.min(data['Temperature'])

print("Average Temperature:", avg_temp)
print("Max Temperature:", max_temp)
print("Min Temperature:", min_temp)
```

```python
print("Missing values:\n", data.isnull().sum())
```

```python
data = data.fillna(0)   # replaces missing with 0
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(data['Date'], data['Temperature'], color='blue', label='Temperature')
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Trend")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data['Date'], data['Rainfall'], color='skyblue')
plt.xlabel("Date")
plt.ylabel("Rainfall (mm)")
plt.title("Daily Rainfall Distribution")
plt.xticks(rotation=45)  # Rotate date labels for better view
plt.show()
```

```python
condition_counts = data['Condition'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Weather Condition Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(data['Date'], data['Temperature'], color='blue', marker='o', linestyle='-')
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Trend")
plt.xticks(rotation=45)   # Rotate x-axis labels for better readability
plt.grid(True)            # Add grid lines
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data['Date'], data['Rainfall'], color='skyblue')
plt.xlabel("Date")
plt.ylabel("Rainfall (mm)")
plt.title("Daily Rainfall")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Date'], data['Temperature'], label="Temperature (°C)", color="red")
plt.plot(data['Date'], data['Humidity'], label="Humidity (%)", color="blue")
plt.xlabel("Date")
plt.ylabel("Values")
plt.title("Temperature vs Humidity")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(data['Temperature'], data['Humidity'], color="green", alpha=0.6)
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.title("Scatter Plot: Temperature vs Humidity")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation between numeric columns
corr = data[['Temperature', 'Rainfall', 'Humidity']].corr()

# Plot heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
